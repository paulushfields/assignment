```python
import pandas as pd
```

```python

file_names = [
    'Household numbers by income band_ext2_clusters',
    'Population_ext2_clusters',
    'Consumer spending by product_ext2_clusters'
]

for file_item in file_names:
    df_data = pd.read_csv("data/" + file_item + ".csv",
                        sep=";",
                        encoding="utf8")
    df_data["countryid"] = ''.join([i for i in df_data['geographyid'][1] if not i.isdigit()])

    df_grouped = df_data.groupby(by=['cluster', 'countryid']).mean()
    # Back to data frame
    df_sum = df_grouped.reset_index()

    print(df_sum)
```
